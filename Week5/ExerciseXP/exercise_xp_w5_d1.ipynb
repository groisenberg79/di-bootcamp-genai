## Answer:

Data analysis is the systematic process of cleaning, transforming, and modeling raw data to discover useful information, suggest conclusions, and support informed decision-making. It involves moving from a "what happened"understanding of the past to a "what should we do" strategy for the future.

In an increasingly data-heavy world, data analysis is vital for moving beyond intuition to evidence-based strategy.

- Informed Decision-Making: It removes guesswork by providing concrete facts and patterns to guide complex choices.

- Operational Efficiency: By identifying bottlenecks and waste,  organizations can streamline workflows and reduce costs.

- Risk Mitigation: Analyzing historical trends helps predict
potential failures or threats (like fraud) before they escalate.

- Competitive Edge: Companies that can quickly interpret market shifts or customer needs can adapt faster than their rivals.

**Common Applications**:
- Healthcare (Disease Diagnosis & Prevention): Analysts use medical records and genetic data to identify at-risk patients and predict disease outbreaks. For example, Google's LYNA tool uses machine learning to identify metastatic breast cancer tumors with 99% accuracy.

- Finance (Fraud Detection): Banks and fintech companies use real-time transaction analysis to spot anomalies that signal fraud. PayPal, for instance, employs AI-driven algorithms to analyze millions of transactions instantly to prevent unauthorized activity.

- Logistics (Route Optimization): Delivery companies use data on traffic, weather, and historical patterns to find the most efficient paths. UPS's ORION system reportedly saves the company 100 million miles and 10 million gallons of fuel annually through these optimizations.

## Answer:

````python
import matplotlib.pyplot as plt
import pandas as pd

df_iris = pd.read_csv("/Users/gabrielroisenbergrodrigues/Desktop/di-bootcamp-genai/Week5/ExerciseXP/data/Iris.csv")

petal_length_mean = df_iris['PetalLengthCm'].mean().round(2)

print(f"The mean of the PetalLengthCm column is: {petal_length_mean}")

petal_length_median = df_iris['PetalLengthCm'].median()

print(f"The median of the PetalLengthCm column is: {petal_length_median}")

petal_length_mode = df_iris['PetalLengthCm'].mode()

print(f"The mode of the PetalLengthCm column is: {petal_length_mode[0]}")
print()

df_iris['PetalLengthCm'].hist()
plt.title('Histogram of Iris Petal Lentgh')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
````
