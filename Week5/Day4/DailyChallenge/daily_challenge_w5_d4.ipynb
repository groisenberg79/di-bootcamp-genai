````python
# Load an image using PIL
image_path = 'flowers/flowers/19_010.png'
original_image = Image.open(image_path)

# Display the original image using matplotlib
plt.imshow(original_image)
````

````python


def rotate_image_30_degrees(image):
  return rotate(original_image, 30, reshape=False )

rotated_image = rotate_image_30_degrees(original_image)

print("Original Image:")
plt.imshow(original_image)
plt.show()

print("Rotated Image:")
plt.imshow(rotated_image)
plt.show()
````

````python
# Horizontal flip (mirror)
h_flipped = ImageOps.mirror(original_image)

print("Original Image:")
plt.imshow(original_image)
plt.show()

print("Horizontal Flipped Image:")
plt.imshow(h_flipped)
plt.show()

# Vertical flip
v_flipped = ImageOps.flip(original_image)

print("Original Image:")
plt.imshow(original_image)
plt.show()

print("Vertical Flipped Image:")
plt.imshow(v_flipped)
plt.show()
````

````python
width, height = original_image.size

zoomed_image = original_image.resize(
    (int(width * 1.2), int(height * 1.2))
)

print("Original Image:")
plt.imshow(original_image)
plt.show()

print("Zoomed Image:")
plt.imshow(zoomed_image)
plt.show()
````
