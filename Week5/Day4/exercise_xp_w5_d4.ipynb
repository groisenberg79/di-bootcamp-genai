# ðŸŒŸ Exercise 2: Transformation Exercise

For each of the following unstructured data sources, propose a method to convert it into structured data. Explain your reasoning.

- A series of blog posts about travel experiences.

  **Answer**: *Blog posts are unstructured text. They can be converted into structured data by extracting key fields such as author, publication date, location, main topics, and sentiment.
  Each blog post can become one row in a table, with columns representing these attributes, making it easier to analyze trends in destinations or traveler opinions.*

- Audio recordings of customer service calls.

  **Answer**: *Audio recordings are unstructured data. The first step is to transcribe the audio into text. From the transcriptions, relevant fields such as customer ID, call duration, issue category, and resolution status can be extracted and stored in a structured database for reporting and analysis.*

- Handwritten notes from a brainstorming session.

  **Answer**: *Handwritten notes can be digitized using optical character recognition (OCR). Once converted to text, ideas can be categorized by topic, author, or priority, and stored in a structured format such as a table or spreadsheet.*

- A video tutorial on cooking.

  **Answer**: *A video tutorial is unstructured multimedia data. It can be transformed by extracting structured elements such as recipe name, ingredients, cooking steps, and cooking time. These elements can then be organized into a structured recipe database.*
